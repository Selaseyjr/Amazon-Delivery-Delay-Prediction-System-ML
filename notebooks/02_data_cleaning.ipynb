```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/amazon_ecommerce_1M.csv")
```

```python
df = df.sample(n=20000, random_state=42)
```

```python
columns_to_keep = [
    "category",
    "subcategory",
    "brand",
    "price",
    "discount",
    "final_price",
    "rating",
    "review_count",
    "stock",
    "seller_rating",
    "shipping_time_days",
    "location",
    "payment_method",
    "delivery_status"
]

df = df[columns_to_keep]
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["delivery_status"].value_counts()
```

Converting to Binary

```python
df["delivery_status"] = df["delivery_status"].apply(
    lambda x: 1 if x == "Delayed" else 0
)
```

```python
df.to_csv("../data/processed/cleaned_delivery_data.csv", index=False)
```
