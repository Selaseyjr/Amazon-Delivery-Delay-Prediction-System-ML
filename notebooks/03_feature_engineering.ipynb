```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/cleaned_delivery_data.csv")
```

Feature 1- High rating

```python
df["high_rating"] = df["rating"].apply(
    lambda x: 1 if x >= 4 else 0
)
```

Feature 2- Fast Shipping

```python
df["fast_shipping"] = df["shipping_time_days"].apply(
    lambda x: 1 if x <= 3 else 0
)
```

Feature 3- Popular product

```python
df["popular_product"] = df["review_count"].apply(
    lambda x: 1 if x >= 100 else 0
)
```
